```python
load_dotenv()
```

```python
api_key = os.getenv("api_key")
environment = os.getenv("environment")
```

```python
pc=Pinecone(
    api_key=api_key,      
    environment=environment       
)
print("Connected to Pinecone")
```

```python
index_name = "diabetes-assistant-index"
```

```python
if index_name not in pc.list_indexes().names():
    pc.create_index(
        name=index_name,
        dimension=384,  
        metric="cosine",
        spec=ServerlessSpec(cloud="aws", region="us-east-1")
    )
    print(f"Index '{index_name}' created successfully.")
```

```python
index = pc.Index(index_name)
```

```python
input_folder = r"D:\AI\Internship\AI Agent Diabetes Diagnosis\AI-Agent-Diabetes-Diagnosis\Datasets\Input"
```

```python
model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
for file_name in os.listdir(input_folder):
    if file_name.endswith(".json"):
        file_path = os.path.join(input_folder, file_name)
        with open(file_path, "r", encoding="utf-8") as f:
            text = f.read()

        vector = model.encode(text).tolist()  
        index.upsert([(file_name, vector, {"text": text})])

print("All JSON files uploaded to Pinecone successfully.")
```
